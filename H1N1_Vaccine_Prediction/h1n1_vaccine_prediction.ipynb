```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
columns=['age_bracket', 'qualification',
       'race', 'sex', 'income_level', 'marital_status', 'housing_status',
       'employment', 'census_msa']
for i in columns:
    df[i]=label_encoder.fit_transform(df[i])
pd.set_option("display.max_columns",None)
df.head(5)  
for i in columns:
    print(df[i].unique())
```

```python
df.drop(["unique_id"],axis=1,inplace=True)
```

```python
df.head(5)
```

```python
corr=df.corr()
sns.heatmap(corr,annot=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X=df.drop(["h1n1_vaccine"],axis=1)
y=df["h1n1_vaccine"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
model_1=LogisticRegression(max_iter=200,C=5)
model_1.fit(X_train,y_train)
model_1.score(X_train,y_train)
```

```python
print(model_1.score(X_test,y_test))
prediction_1=model_1.predict(X_test)
```

```python
from sklearn.svm import SVC
model_2=SVC()
model_2.fit(X_train,y_train)
model_2.score(X_train,y_train)
```

```python
print(model_2.score(X_test,y_test))
prediction_2=model_2.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_3=DecisionTreeClassifier(max_depth=9,min_samples_leaf=3,criterion="gini")
model_3.fit(X_train,y_train)
model_3.score(X_train,y_train)
```

```python
print(model_3.score(X_test,y_test))
prediction_3=model_3.predict(X_test)
```

```python
from sklearn.ensemble import BaggingClassifier
model_4=BaggingClassifier(n_estimators=6,random_state=1,max_samples=0.2)
model_4.fit(X_train,y_train)
model_4.score(X_train,y_train)
```

```python
print(model_4.score(X_test,y_test))
prediction_4=model_4.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_5=RandomForestClassifier(n_estimators=50,max_depth=12,min_samples_split=5,min_samples_leaf=2)
model_5.fit(X_train,y_train)
model_5.score(X_train,y_train)
```

```python
print(model_5.score(X_test,y_test))
prediction_5=model_5.predict(X_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
model_6=AdaBoostClassifier()
model_6.fit(X_train,y_train)
model_6.score(X_train,y_train)
```

```python
print(model_6.score(X_test,y_test))
prediction_6=model_6.predict(X_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_7=GradientBoostingClassifier(learning_rate=0.75)
model_7.fit(X_train,y_train)
model_7.score(X_train,y_train)
```

```python
print(model_7.score(X_test,y_test))
prediction_7=model_7.predict(X_test)
```

```python
## We want do the confusion matrix calculations for the skewd_data
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,prediction_1))
print(confusion_matrix(y_test,prediction_2))
print(confusion_matrix(y_test,prediction_3))
print(confusion_matrix(y_test,prediction_4))
print(confusion_matrix(y_test,prediction_5))
print(confusion_matrix(y_test,prediction_6))
print(confusion_matrix(y_test,prediction_7))
```

```python
from sklearn import metrics
from sklearn.metrics import f1_score
##Classification Report of the samples
print(metrics.classification_report(y_test,prediction_1))
print(metrics.classification_report(y_test,prediction_2))
print(metrics.classification_report(y_test,prediction_3))
print(metrics.classification_report(y_test,prediction_4))
print(metrics.classification_report(y_test,prediction_5))
print(metrics.classification_report(y_test,prediction_6))
print(metrics.classification_report(y_test,prediction_7))
##f1_score calculations of the metrics
##Want to calculate the highest f1score
print(f1_score(y_test,prediction_1))
print(f1_score(y_test,prediction_2))
print(f1_score(y_test,prediction_3))
print(f1_score(y_test,prediction_4))
print(f1_score(y_test,prediction_5))
print(f1_score(y_test,prediction_6))
print(f1_score(y_test,prediction_7))
```

```python
##Hence we are getting the more f1_score using the Gradient_Boosting_Classifier hence it is the best model
##Hence the model is the 7th model
cm=metrics.confusion_matrix(y_test, prediction_7, labels=[1, 0])

df_cm = pd.DataFrame(cm, index = [i for i in ["1","0"]],
                  columns = [i for i in ["Predict 1","Predict 0"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True,fmt='g')
##This is calculated for only the test samples
```
